```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.select import Select
from scraperutils import *
from datamodel import Fields
import time
```

```python
def blah_blah(caseID, countryName, year, name, url):
    project_details = dict();
    driver.get(url);
    time.sleep(3);

    YEAR = year;

    countryElem = driver.find_element_by_css_selector("span[id='projectInformationCountry']");
    COUNTRY = driver.execute_script("return arguments[0].innerHTML", countryElem);

    PROJECT_NAME = name;


    projectElem = driver.find_element_by_css_selector("span[id='projectInformationID']");
    PROJECT_ID = driver.execute_script("return arguments[0].innerHTML", projectElem);

    PROJECT_NUMBER = caseID;

    RELATED_PROJECT_NUMBER = None;

    PROJECT_TYPE = None;

    #Wrong locator in the original scrapper. Keeping it a TODO for future.
    PROJECT_LOAN_AMOUNT = None;

    #Wrong locator in the original scrapper. Keeping it a TODO for future.
    SECTOR = None;

    #Wrong locator in the original scrapper. Keeping it a TODO for future.
    ISSUES = None;

    try: 
        FILERS = driver.find_element_by_xpath('//*[@id="caseInformationRequesters"]/div/p').text
    except NoSuchElementException:
        FILERS = driver.find_element_by_id('caseInformationRequesters').text


    envCatElem = driver.find_element_by_css_selector("span[id='projectInformationCategory']");
    ENVIRONMENTAL_CATEGORY = driver.execute_script("return arguments[0].innerHTML", envCatElem);

        # link to the project
    HYPERLINK = url;



    stages = driver.find_elements_by_xpath('//*[@id="CaseViewTabs"]/ul/li')

    for stage in range(1, len(stages)+1):
        each_stage = driver.find_element_by_xpath('//*[@id="CaseViewTabs"]/ul/li[%s]' % stage)
        stage_class = each_stage.get_attribute('class')
        stage_text = each_stage.text
        for i in stage_text:
            if stage_text == 'Notice of Receipt':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                    FILING_DATE = 'Completed'
                else: 
                    FILING_DATE = None
            if stage_text == 'Registration of Request':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                    REGISTRATION_START_DATE = 'Completed'
                    REGISTRATION_END_DATE = 'Completed'
                else: 
                    REGISTRATION_START_DATE = None
                    REGISTRATION_END_DATE = None
            if stage_text in ('Management Response', 'Eligibility Report'):
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                    ELIGIBILITY_START_DATE = 'Completed'
                else:
                    ELIGIBILITY_START_DATE = None
            if stage_text == 'Board Approval':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                    ELIGIBILITY_END_DATE = 'Completed'
                else: 
                    ELIGIBILITY_END_DATE = None
            if stage_text in ('Investigation Ongoing', 'Investigation Report', 'Management Report'):
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                     COMPLIANCE_REVIEW_START_DATE = 'Completed'
                else: 
                     COMPLIANCE_REVIEW_START_DATE = None
            if stage_text == 'Investigation Report':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                     IS_COMPLIANCE_REPORT_ISSUED = 'TRUE'
                else: 
                     IS_COMPLIANCE_REPORT_ISSUED = 'FALSE'
            if stage_text == 'Board Discussion':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                     COMPLIANCE_REVIEW_END_DATE = 'Completed'
                else: 
                     COMPLIANCE_REVIEW_END_DATE = None
            if stage_text == 'Follow up and Progress Report':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                    MONITORING_START_DATE = 'Completed'
                    MONITORING_END_DATE = 'Completed'
                else: 
                    MONITORING_START_DATE = None
                    MONITORING_END_DATE = None
            if stage_text == 'Process Completed':
                if stage_class in ('completedCurrent', 'completedCurrent active', 'liClassFirst completedCurrent'):
                    DATE_CLOSED = 'Completed'
                    COMPLAINT_STATUS = 'Closed'
                else: 
                    DATE_CLOSED = None
                    COMPLAINT_STATUS = 'Open'

    DISPUTE_RESOLUTION_START_DATE = None;
    DISPUTE_RESOLUTION_END_DATE = None;


    documents_list = driver.find_elements_by_class_name('DocsList')
    attached_documents = "";
    for document in documents_list:
        attached_documents = attached_documents + ", " + document.find_element_by_css_selector('a').get_attribute('href')

    IAM = 'Panel';
    IAM_ID = 23;

    
    project_details[Fields.IAM.name] = IAM;
    project_details[Fields.IAM_ID.name] = IAM_ID;
    project_details[Fields.YEAR.name] = YEAR;
    project_details[Fields.COUNTRY.name] = COUNTRY;
    project_details[Fields.PROJECT_NAME.name] = PROJECT_NAME;
    project_details[Fields.PROJECT_ID.name] = PROJECT_ID;
    project_details[Fields.PROJECT_NUMBER.name] = PROJECT_NUMBER;
    project_details[Fields.RELATED_PROJECT_NUMBER.name] = RELATED_PROJECT_NUMBER;
    project_details[Fields.PROJECT_TYPE.name] = PROJECT_TYPE;
    project_details[Fields.PROJECT_LOAN_AMOUNT.name] = PROJECT_LOAN_AMOUNT;
    project_details[Fields.SECTOR.name] = SECTOR;
    project_details[Fields.ISSUES.name] = ISSUES;
    project_details[Fields.FILERS.name] = FILERS;
    project_details[Fields.FILING_DATE.name] = FILING_DATE;
    project_details[Fields.ENVIRONMENTAL_CATEGORY.name] = ENVIRONMENTAL_CATEGORY;
    project_details[Fields.COMPLAINT_STATUS.name] = COMPLAINT_STATUS;
    project_details[Fields.DATE_CLOSED.name] = DATE_CLOSED;
    project_details[Fields.HYPERLINK.name] = HYPERLINK;

        
#     print("#PART-I");
#     print(IAM);
#     print(IAM_ID);
#     print(YEAR);
#     print(COUNTRY);
#     print(PROJECT_NAME);
#     print(PROJECT_ID);
#     print(PROJECT_NUMBER);
#     print(RELATED_PROJECT_NUMBER);
#     print(PROJECT_TYPE);
#     print(PROJECT_LOAN_AMOUNT);
#     print(SECTOR);
#     print(ISSUES);
#     print(FILERS);
#     print(FILING_DATE);
#     print(ENVIRONMENTAL_CATEGORY);
#     print(COMPLAINT_STATUS);
#     print(DATE_CLOSED);
#     print(HYPERLINK);

    project_details[Fields.REGISTRATION_START_DATE.name] = REGISTRATION_START_DATE;
    project_details[Fields.REGISTRATION_END_DATE.name] = REGISTRATION_END_DATE;
    project_details[Fields.ELIGIBILITY_START_DATE.name] = ELIGIBILITY_START_DATE;
    project_details[Fields.ELIGIBILITY_END_DATE.name] = ELIGIBILITY_END_DATE;
    project_details[Fields.REGISTRATION_START_DATE.name] = REGISTRATION_START_DATE;
    project_details[Fields.DISPUTE_RESOLUTION_START_DATE.name] = DISPUTE_RESOLUTION_START_DATE;
    project_details[Fields.DISPUTE_RESOLUTION_END_DATE.name] = DISPUTE_RESOLUTION_END_DATE;
    project_details[Fields.COMPLIANCE_REVIEW_START_DATE.name] = COMPLIANCE_REVIEW_START_DATE;
    project_details[Fields.COMPLIANCE_REVIEW_END_DATE.name] = COMPLIANCE_REVIEW_END_DATE;
    project_details[Fields.MONITORING_START_DATE.name] = MONITORING_START_DATE;
    project_details[Fields.MONITORING_END_DATE.name] = MONITORING_END_DATE;
    project_details[Fields.IS_COMPLIANCE_REPORT_ISSUED.name] = IS_COMPLIANCE_REPORT_ISSUED;

#     print("#PART-II");
#     print(REGISTRATION_START_DATE);
#     print(REGISTRATION_END_DATE);
#     print(ELIGIBILITY_START_DATE);
#     print(ELIGIBILITY_END_DATE);
#     print(REGISTRATION_START_DATE);
#     print(REGISTRATION_END_DATE);
#     print(DISPUTE_RESOLUTION_START_DATE);
#     print(DISPUTE_RESOLUTION_END_DATE);
#     print(COMPLIANCE_REVIEW_START_DATE);
#     print(COMPLIANCE_REVIEW_END_DATE);
#     print(MONITORING_START_DATE);
#     print(MONITORING_END_DATE);
#     print(IS_COMPLIANCE_REPORT_ISSUED);

    
    #project_details[Fields.attached_documents.name] = attached_documents;
    
#     print("#PART-III");
#     print(attached_documents);
    

    return project_details;
```
